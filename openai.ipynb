## Image Generation via Prompt:


```python
# Define the prompt for generating an image
PROMPT = "An eco-friendly computer from the 90s in the style of vaporwave"

# Create an image using the prompt
response = openai.Image.create(
    prompt=PROMPT,
    n=1,
    size="256x256",
)

# Print the URL of the generated image
print(response["data"][0]["url"])
```

```python
def generate(text):
  res = openai.Image.create(
    prompt=text,
    n=1,
    size="512x512",
  )
  return res["data"][0]["url"]

Image.open(requests.get(generate('un guerrero medieval'), stream=True).raw)
```

```python
# gpt-4

# Prompt the user to input their message
gpt_prompt = input("Enter your prompt: ")
print(gpt_prompt)

# Create a message in the specified format
message = [{"role": "user", "content": gpt_prompt}]

# Generate a response using the GPT-4 model
response = openai.ChatCompletion.create(
    model="gpt-4",
    messages=message,
    temperature=0.2,
    max_tokens=1000,
    frequency_penalty=0.0
)

# Print the response
print(response)
```
